# Cleaning


```python
backup=df.copy()
```

```python
backup.info()
```

```python
df.info()
```

```python
# only those records are taken whose value is numeric
df=df[df['year'].str.isnumeric()]
```

```python
# convert year from object to int
df['year']=df['year'].astype('int')
```

```python
#exclude ask for price
df=df[df['Price']!='Ask For Price']
```

```python
# convert price from object 
df['Price']=df['Price'].str.replace(',','').astype('int')
```

```python
# removing kms
df['kms_driven']=df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
# removing petrol
df=df[df['kms_driven']!='Petrol']
```

```python
# converting kms driven to int
df['kms_driven']=df['kms_driven'].astype('int')
```

```python
# removing nan values
df=df[~df['fuel_type'].isna()]
df['fuel_type'].isna().sum()
```

```python
# taking first three words of name 
len=df['name'].size
len
```

```python
df['name']=df['name'].str.split(',').str.slice(0,3).str.join('')
```

# Model

```python
X= df.drop(columns='Price')
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
ohe=OneHotEncoder()
encoded=ohe.fit_transform(X[['name','company','fuel_type']])
```

```python
encoded.toarray()
```

```python
# Part	                        Meaning
# make_column_transformer(...)	-> Creates a pipeline to apply different preprocessing steps to different columns
# (OneHotEncoder(), ['name', 'company', 'fuel_type'])-> 	Applies OneHotEncoder to these 3 categorical columns
# remainder='passthrough'	-> Keeps all other columns as-is (not transformed)

col_transformer=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe=make_pipeline(
    col_transformer,
    lr
)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_predict=pipe.predict(X_test)
```

```python
r2_score(y_test,y_predict)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    ohe=OneHotEncoder()
    encoded=ohe.fit_transform(X[['name','company','fuel_type']])
    col_transformer=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
    pipe=make_pipeline(col_transformer,lr)
    pipe.fit(X_train,y_train)
    y_predict=pipe.predict(X_test)
    print(r2_score(y_test,y_predict),i)
    scores.append(r2_score(y_test,y_predict))
    
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.argmax(scores))
ohe=OneHotEncoder()
encoded=ohe.fit_transform(X[['name','company','fuel_type']])
col_transformer=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
pipe=make_pipeline(col_transformer,lr)
pipe.fit(X_train,y_train)
y_predict=pipe.predict(X_test)
print(r2_score(y_test,y_predict))
```
